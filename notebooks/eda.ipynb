# üîç **An√°lisis Exploratorio de Datos**

### üìö **Librer√≠as**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### üìÇ **Carga de datos**

```python
df = pd.read_csv('../data/df_limpio.csv')

df.head(3)
```

### üìê **An√°lisis descriptivo**

```python
print("Variables num√©ricas: ")
df.describe().T
```

```python
print("Variables categ√≥ricas: ")
df.describe(include="object", exclude="number").T
```

### üîç **An√°lisis de la variable Churn**

**An√°lisis univariado**

```python
# Tasa de Churn general
print(round(df['Churn'].value_counts(normalize=True) * 100, 2))
```

```python
churn = df['Churn'].value_counts()

sns.barplot(x=churn.index, y=churn.values)

plt.title('Distribuci√≥n de Churn')
plt.show()
```

**An√°lisis bivariado y multivariado**

```python
sns.countplot(data=df, x="G√©nero", hue="Churn")
plt.title("Churn por g√©nero")
plt.ylabel("Cantidad de clientes")

plt.show()
```

```python
sns.countplot(data=df, x="Adulto Mayor", hue="Churn")
plt.title("Churn por Adulto Mayor (1=Adulto Mayor)")
plt.ylabel("Cantidad de clientes")

plt.show()
```

```python
sns.countplot(data=df, x="Pareja", hue="Churn")
plt.title("Churn por estado civil (Yes= tiene pareja)")
plt.ylabel("Cantidad de clientes")

plt.show()
```

```python
sns.countplot(data=df, x="Dependientes", hue="Churn")
plt.title("Churn por Dependientes (Yes=tiene dependientes)")
plt.ylabel("Cantidad de clientes")

plt.show()
```

```python
df['M√∫ltiples L√≠neas'].value_counts()
```

```python
sns.countplot(data=df, x='M√∫ltiples L√≠neas', hue="Churn")
plt.title("Churn por M√∫ltiples L√≠neas")
plt.ylabel("Cantidad de clientes")

plt.show()
```

```python
sns.countplot(data=df, x='Respaldo Online', hue="Churn")
plt.title("Churn por Respaldo Online")
plt.ylabel("Cantidad de clientes")

plt.show()
```

```python
sns.countplot(data=df, x='Soporte T√©cnico', hue="Churn")
plt.title("Churn por Soporte T√©cnico")
plt.ylabel("Cantidad de clientes")

plt.show()
```

```python
sns.countplot(data=df, x='Tipo de Contrato', hue="Churn")
plt.title("Churn por Tipo de Contrato")
plt.ylabel("Cantidad de clientes")

plt.show()
```

**Conteo de evasi√≥n por variables num√©ricas**

```python
sns.boxplot(data=df, x="Churn", y="Cobro Total")

plt.title("Boxplot de Cobro Total por Churn")
plt.show()
```

```python
sns.boxplot(data=df, x="Churn", y="Cobro Mensual")

plt.title("Boxplot de Cobro Mensual por Churn")
plt.show()
```

```python
sns.boxplot(data=df, x="Churn", y="Tiempo de Servicio (meses)")

plt.title("Boxplot de Tiempo de Servicio (meses) por Churn")
plt.show()
```
